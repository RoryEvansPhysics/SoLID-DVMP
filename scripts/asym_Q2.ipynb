# Plot DVMP Bining Result

```python
### Load Python Lib##### 
########
import numpy as np
from math import *
import matplotlib.pyplot as plt
from random import gauss, randint
from matplotlib import rc
from matplotlib.ticker import FixedLocator, MultipleLocator, FormatStrFormatter

#very import to enable this so the plots can be showed in the page
get_ipython().magic(u'matplotlib inline')

# Use LaTeX font.
plt.rc('text', usetex=True)
plt.rc('font',**{'family':'serif','serif':['Computer Modern Roman'],'size':20})
```

## Read in Data File

### Read in results in 1st Q2 bin

```python
## Reading data file
Q2 = np.zeros(7, dtype=float)
x = np.zeros(7, dtype=float)
W = np.zeros(7, dtype=float)
t = np.zeros(7, dtype=float)
R = np.zeros(7, dtype=float)
F = np.zeros(7, dtype=float)
xs = np.zeros(7, dtype=float)
Asym = np.zeros(7, dtype=float)
dAsym = np.zeros(7, dtype=float)
N = np.zeros(7, dtype=float)
N_raw = np.zeros(7, dtype=float)

Q2_avg=0.0
W_avg=0.0
N_Total = 0.0
N_raw_Total = 0.0

#prd = "_prd"
prd = ""

#com = "w/ PRD"
#com = "w/o PRD"
com=""

filename="noFR_n_0_new1.dat"
lines = open(filename, 'r').readlines()
for i in range(0,len(lines)):
    values = lines[i].split()    
    Q2[i] = float(values[1])
    x[i] = float(values[2])
    W[i] = float(values[3])
    t[i] = float(values[4])
    R[i] = float(values[5])
    F[i] = float(values[6])
    xs[i] = float(values[7])
    Asym[i] = float(values[8])
    dAsym[i] = float(values[9])
    N[i] = float(values[10])
    N_raw[i] = float(values[11])

    Q2_avg += Q2[i] * N[i]
    W_avg += W[i] * N[i]
    N_Total += N[i]
    N_raw_Total += N_raw[i]
    
Q2_avg /= N_Total
W_avg /= N_Total
print "N_Total_Raw=%d, N_Total=%d"%(N_raw_Total, N_Total)
```

## Read in results in 2nd Q2 bin

```python
## Reading data file
Q2_prd = np.zeros(7, dtype=float)
x_prd = np.zeros(7, dtype=float)
W_prd = np.zeros(7, dtype=float)
t_prd = np.zeros(7, dtype=float)
R_prd = np.zeros(7, dtype=float)
F_prd = np.zeros(7, dtype=float)
xs_prd = np.zeros(7, dtype=float)
Asym_prd = np.zeros(7, dtype=float)
dAsym_prd = np.zeros(7, dtype=float)
N_prd = np.zeros(7, dtype=float)
N_raw_prd = np.zeros(7, dtype=float)

Q2_avg_prd=0.0
W_avg_prd=0.0
N_Total_prd = 0.0
N_raw_Total_prd = 0.0

#filename_prd="proton_prd_solid.dat"
filename_prd="noFR_n_0_new2.dat"
lines = open(filename_prd, 'r').readlines()
for i in range(0,len(lines)):
    values = lines[i].split()    
    Q2_prd[i] = float(values[1])
    x_prd[i] = float(values[2])
    W_prd[i] = float(values[3])
    t_prd[i] = float(values[4])
    R_prd[i] = float(values[5])
    F_prd[i] = float(values[6])
    xs_prd[i] = float(values[7])
    Asym_prd[i] = float(values[8])
    dAsym_prd[i] = float(values[9])
    N_prd[i] = float(values[10])
    N_raw_prd[i] = float(values[11])

    Q2_avg_prd += Q2_prd[i] * N_prd[i]
    W_avg_prd += W_prd[i] * N_prd[i]
    N_Total_prd += N_prd[i]
    N_raw_Total_prd += N_raw_prd[i]
    
Q2_avg_prd /= N_Total_prd
W_avg_prd /= N_Total_prd
print "N_Total_Raw=%e, N_Total=%e"%(N_raw_Total+N_raw_Total_prd, N_Total+N_Total_prd)
```
